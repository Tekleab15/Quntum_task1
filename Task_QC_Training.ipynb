```python
# Import required libraries
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit import transpile
import matplotlib.pyplot as plt
```

# **Create a Quantum Circuit with a Hadamard Gate**

- Write a Qiskit program to create a single-qubit circuit.
- Apply a Hadamard gate to the qubit.
- Measure the qubit and visualize the result using a histogram.

```python
# Step 1: Create a Quantum Circuit with 1 qubit and 1 classical bit
circuit = QuantumCircuit(1, 1)
# Step 2: Apply Hadamard Gate to the qubit
circuit.h(0)
# Step 3: Measure the qubit
circuit.measure(0, 0)
print("Quantum Circuit:")
print(circuit.draw())

# Step 4: Simulate the Circuit
simulator = AerSimulator()
compiled_circuit = transpile(circuit, simulator)
result = simulator.run(compiled_circuit, shots=1024).result()
counts = result.get_counts()

# Step 5: Plot the Results
print("Measurement Results:", counts)

# Ensure histogram displays correctly
fig, ax = plt.subplots(figsize=(8, 6))
plot_histogram(counts, ax=ax)
plt.title("Measurement Results")
plt.show()
```

# **Implement a Quantum Coin Flip**
- Simulate a fair coin flip using a quantum circuit.
- Print the probability of getting Heads or Tails.

```python
# Simulate the circuit
simulator = AerSimulator()
compiled_circuit = transpile(circuit, simulator)
result = simulator.run(compiled_circuit, shots=1024).result()
counts = result.get_counts()

# Plot the results
print("Coin Flip Results:", counts)

# Calculate the probabilities
total_shots = sum(counts.values())
prob_heads = counts.get('0', 0) / total_shots
prob_tails = counts.get('1', 0) / total_shots

# Print the results
print("Coin Flip Results:", counts)
print(f"Probability of Heads: {prob_heads * 100:.2f}%")
print(f"Probability of Tails: {prob_tails * 100:.2f}%")

# Ensure histogram displays correctly
fig, ax = plt.subplots(figsize=(8, 6))
plot_histogram(counts, ax=ax)
plt.title("Coin Flip Results")
plt.show()
```

# **Implement a Bell State (Entanglement)**
- Create a two-qubit circuit.
- Apply a Hadamard gate to the first qubit and a CNOT gate between the two qubits.
- Measure both qubits and verify if they are always correlated.

```python
circuit = QuantumCircuit(2, 2)

# Apply a Hadamard gate to the first qubit
circuit.h(0)

# Apply a CNOT gate between the first and second qubits
circuit.cx(0, 1)

# Measure both qubits
circuit.measure([0, 1], [0, 1])

# Visualize the circuit
print("Quantum Circuit for Bell State:")
print(circuit.draw())

# Simulate the circuit
simulator = AerSimulator()
compiled_circuit = transpile(circuit, simulator)
result = simulator.run(compiled_circuit, shots=1024).result()
counts = result.get_counts()
```

# **Theory Checkins**

### **How does superposition affect the information processing capabilities of a quantum computer?**

*Superposition* allows quantum bits (qubits) to exist in multiple states simultaneously, unlike classical bits which are either 0 or 1. This enables quantum computers to process a vast amount of information in parallel, potentially solving complex problems much faster than classical computers.*

###How do quantum gates manipulate interference to solve problems faster than classical computers?

*Quantum gates manipulate the probabilities of qubit states using interference. By carefully controlling the phases of qubits, quantum algorithms can amplify the probability of correct answers while canceling out incorrect ones, leading to faster problem-solving capabilities compared to classical approaches.*

### How does measurement collapse a quantum state, and why is it important in quantum computing algorithms?

Measurement collapses a qubit's superposition state into one of the basis states (0 or 1). This collapse is crucial in quantum computing algorithms because it provides a definitive outcome from a probabilistic quantum state, allowing us to extract meaningful and useful results from quantum computations.
